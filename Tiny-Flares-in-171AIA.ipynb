# AIA 171

```python
from sunpy.map import MapCubed
from sunpy.map import Map
import glob
import astropy.units as u
from sunpy.time import parse_time
from matplotlib.patches import Rectangle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
data_dir = '/Volumes/500GB/AIA/'
str_indices = {'94':'', '131':'', '171':'', '193':'', '211':'', '304':'', '335':'', '1600':'', '1700':''}
file_list = {}
maps = {}
for key in str_indices:
    file_list.update({key: glob.glob(data_dir + '*' + key + '*.fits')})
    
these_files = file_list['171'][49:90]

maps = []

for f in these_files:
    maps.append(Map(f))

cubed = MapCubed(maps)
```

## Full resolution

```python
x_range = u.Quantity([-900, 0], 'arcsec')
y_range = u.Quantity([-700, 200], 'arcsec')
scubed = cubed.submap(x_range, y_range)
scubed[0].plot()
plt.colorbar()
```

## Rebinning [15 pix]

```python
sscubed = scubed.superpixel(u.Quantity([15 * u.pixel, 15 * u.pixel]))
sscubed[0].plot()
plt.colorbar()
```

# Steven's Flare

```python
sscubed[0].plot(vmin=1e5,vmax=4e5)
ax = plt.gca()
xrect = 70
yrect = 58
ax.add_patch(Rectangle((xrect-0.5, yrect-0.5), 2, 2, fill=None,edgecolor='red',linewidth=1))
plt.colorbar()
```

```python
sscubed.std().plot(vmin=1e4,vmax=5e4)
ax = plt.gca()
ax.add_patch(Rectangle((xrect-0.5, yrect-0.5), 2, 2, fill=None,edgecolor='red',linewidth=1))
plt.colorbar()
```

```python
sscubed.lightcurve(yrect,xrect).plot()
```
