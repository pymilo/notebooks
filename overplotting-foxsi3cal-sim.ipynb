### Doing this but using the plot(detector) function

```python
%matplotlib inline
import matplotlib.pyplot as plt
from astropy.io import fits
from foxsisim.detector import Detector
from foxsisim.plotting import plot, scatterHist
from foxsisim.util import load_rays
import numpy as np
```

## 26 arcmin

```python
# SIMULATION
rays26 = load_rays('/Users/Kamilobu/Desktop/FOXSI/FOXSI-3/FOXSI-SIM-ANALYSIS/26.0arcmin1000000.csv')

detector = Detector(width=2.76, #27.6 x 27.6 mm values taken from Steven
                    height=2.76,
                    normal = [0,0,1],
                    center = [0.5,0,230],
                    reso = [1024,1024])

detector.catchRays(rays26)
plot(detector,energy_range=None) # This create detector.pixels
## Cretaing a transparent background to visualize data easier.
det = detector.pixels[:,:,1]
detm = np.ma.masked_where(det != 1, det)
```

```python
# DATA
hdulist = fits.open('FOXSI3Data/Foxsi_mod6_26arcmin_0.5mA.fits')
hdu = hdulist[0]

## Plotting
fig, ax = plt.subplots(figsize=(16,16))
ax.set_title('26 arcmin - Data')
ax.imshow(hdu.data[0,:,:], origin='lower',cmap='Blues',vmin=1600,vmax=2000)
ax.imshow(detm,cmap='autumn')
```

```python
detectorRays = detector.rays 
fig = plt.figure(figsize=(15,15))
scatterHist(detectorRays,fig)
```
