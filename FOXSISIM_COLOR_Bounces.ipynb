# Debugging the color coding of the bounced rays

```python
from foxsisim.module import Module
from foxsisim.detector import Detector
from foxsisim.source import Source
from foxsisim.plotting import scatterHist,plot
import numpy as np
import matplotlib.pylab as plt
```

```python
source_distance = -1e4 ##cm
source = Source(type='point', center=[0, 0, source_distance])
module = Module(radii=[5.,4.5, 4.])
detector = Detector(width=8, 
                    height=8,
                    normal = [0,0,1],
                    center = [0,0,230],
                    reso = [1024,1024])

rays = source.generateRays(module.targetFront, 20)
module.passRays(rays, robust=True)
detector.catchRays(rays)
```

```python
scatterHist(rays)
plt.show()
```
