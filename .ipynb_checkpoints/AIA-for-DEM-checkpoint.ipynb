```python
import seaborn as sns
from datetime import datetime, timedelta

#t0 = '11-Dec-2014 19:11:00.000'
t0 = datetime(2014,12,11,19,11,0)
t1_pos2_start = timedelta(seconds=166.5)
t1_pos2_end = timedelta(seconds=205.0)

ti = t0 + t1_pos2_start
tf = t0 + t1_pos2_end
```

# AIA averages and std for all wavelengths

```python
from sunpy.map import MapCubed
from sunpy.map import Map
import glob
import astropy.units as u
from sunpy.time import parse_time
from matplotlib.patches import Rectangle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
data_dir = '/Volumes/500GB/AIA/'
str_indices = {'94':'', '131':'', '171':'', '193':'', '211':'', '304':'', '335':'', '1600':'', '1700':''}
file_list = {}
maps = {}
for key in str_indices:
    file_list.update({key: glob.glob(data_dir + '*' + key + '*.fits')})

DEM = {'Tiny-Flare1':{'094':{'ave':'','std':''},'131':{'ave':'','std':''},'171':{'ave':'','std':''},
                '193':{'ave':'','std':''},'211':{'ave':'','std':''},'304':{'ave':'','std':''},
                '335':{'ave':'','std':''}},
       'Tiny-Flare2':{'094':{'ave':'','std':''},'131':{'ave':'','std':''},'171':{'ave':'','std':''},
                '193':{'ave':'','std':''},'211':{'ave':'','std':''},'304':{'ave':'','std':''},
                '335':{'ave':'','std':''}},
       'Tiny-Flare3':{'094':{'ave':'','std':''},'131':{'ave':'','std':''},'171':{'ave':'','std':''},
                '193':{'ave':'','std':''},'211':{'ave':'','std':''},'304':{'ave':'','std':''},
                '335':{'ave':'','std':''}},
       'Tiny-Flare4':{'094':{'ave':'','std':''},'131':{'ave':'','std':''},'171':{'ave':'','std':''},
                '193':{'ave':'','std':''},'211':{'ave':'','std':''},'304':{'ave':'','std':''},
                '335':{'ave':'','std':''}},
       'Tiny-Flare5':{'094':{'ave':'','std':''},'131':{'ave':'','std':''},'171':{'ave':'','std':''},
                '193':{'ave':'','std':''},'211':{'ave':'','std':''},'304':{'ave':'','std':''},
                '335':{'ave':'','std':''}},
      }
```

Location of the Tiny Flares:

```python
xrect = [70, 49, 50, 59, 61]
yrect = [58, 54, 17, 48, 14]
```

## AIA-094

```python
these_files = file_list['94'][59:64]
maps = []
for f in these_files:
    maps.append(Map(f))
cubed = MapCubed(maps)
x_range = u.Quantity([-900, 0], 'arcsec')
y_range = u.Quantity([-700, 200], 'arcsec')
scubed = cubed.submap(x_range, y_range)
sscubed094 = scubed.superpixel(u.Quantity([15 * u.pixel, 15 * u.pixel]))
```

```python
dat=[]
for x,y in zip(xrect,yrect):
    dat.append(sscubed094.lightcurve(y,x).data)

for i in range(len(xrect)):    
    DEM['Tiny-Flare'+str(i+1)]['094']['ave'] = dat[i].values.mean()
    DEM['Tiny-Flare'+str(i+1)]['094']['std'] = dat[i].values.std()
```

```python
dat[0]
```

```python
f, ax0 = plt.subplots()
sscubed094.lightcurve(yrect[0],xrect[0]).plot(figure=f,c='#479084')
ax0.legend(['AIA094$\AA$'])
```
