# Light Curves

```python
from sunpy.map import MapCubed
from sunpy.map import Map
import glob
import astropy.units as u
from sunpy.time import parse_time
from matplotlib.patches import Rectangle
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
data_dir = '/Volumes/500GB/AIA/'
str_indices = {'94':'', '131':'', '171':'', '193':'', '211':'', '304':'', '335':'', '1600':'', '1700':''}
file_list = {}
maps = {}
for key in str_indices:
    file_list.update({key: glob.glob(data_dir + '*' + key + '*.fits')})
```

# Steven's Tiny Flare

## AIA-94

```python
these_files = file_list['94'][49:90]
maps = []
for f in these_files:
    maps.append(Map(f))
cubed = MapCubed(maps)
x_range = u.Quantity([-900, 0], 'arcsec')
y_range = u.Quantity([-700, 200], 'arcsec')
scubed = cubed.submap(x_range, y_range)
sscubed094 = scubed.superpixel(u.Quantity([15 * u.pixel, 15 * u.pixel]))
```

```python
sscubed094[5].plot(vmin=5e2,vmax=8e3)
ax = plt.gca()
xrect = 70
yrect = 58
ax.add_patch(Rectangle((xrect-0.5, yrect-0.5), 1, 1, fill=None,edgecolor='red',linewidth=1))
plt.colorbar()
```

```python
f, ax0 = plt.subplots()  
sscubed094.lightcurve(yrect,xrect).plot(figure=f,c='#479084')
ax0.legend(['AIA094$\AA$'])
```
