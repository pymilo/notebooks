# Scatter Plot: Single Bounces in Red and Double Bounces in Green

Legitimate rays that reach the focal plane after one or two bounces have a z-component = 230. as expected. The color coding of these rays needs to be check. Nevertheless, rays which destination has a z-component of **zero** should not rech the focal plane, but they are still plot when we use *Scattering*.

```python
from foxsisim.module import Module
from foxsisim.detector import Detector
from foxsisim.source import Source
from foxsisim.plotting import scatterHist,plot
import numpy as np
import matplotlib.pylab as plt
```

```python
source_distance = -10000
source = Source(type='point', center=[0, 0, source_distance])
module = Module()
detector = Detector(width=8, 
                    height=8,
                    normal = [0,0,1],
                    center = [0,0,230],
                    reso = [1024,1024])

rays = source.generateRays(module.targetFront, 50000)
module.passRays(rays, robust=True)
detector.catchRays(rays)
```

```python
scatterHist(rays)
plt.show()
```
